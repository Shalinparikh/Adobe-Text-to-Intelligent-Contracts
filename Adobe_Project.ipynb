## Dependency Parser

```python
nlp = spacy.load("en_core_web_sm")
def merge_phrases(doc):
    with doc.retokenize() as retokenizer:
        for np in list(doc.noun_chunks):
            attrs = {
                "tag": np.root.tag_,
                "lemma": np.root.lemma_,
                "ent_type": np.root.ent_type_,
            }
            retokenizer.merge(np, attrs=attrs)
    return doc

def merge_punct(doc):
    spans = []
    for word in doc[:-1]:
        if word.is_punct or not word.nbor(1).is_punct:
            continue
        start = word.i
        end = word.i + 1
        while end < len(doc) and doc[end].is_punct:
            end += 1
        span = doc[start:end]
        spans.append((span, word.tag_, word.lemma_, word.ent_type_))
    with doc.retokenize() as retokenizer:
        for span, tag, lemma, ent_type in spans:
            attrs = {"tag": tag, "lemma": lemma, "ent_type": ent_type}
            retokenizer.merge(span, attrs=attrs)
    return doc
```

# **Evaluation Metric**

```python
# Calculate accuracy percentage between two lists
def accuracy_metric(actual, predicted):
  correct = 0
  for i in range(len(actual)):
    if (set(predicted[i]).issubset(set(actual[i]))):
      correct += 1
  return correct / len(actual) * 100
```

```python
def accuracy_metric(actual, predicted):
  correct = []
  for i in range(len(actual)):
    score = 0
    for j in predicted[i]:
      if j in actual[i]:
        score += 1
    if len(actual[i]) > 0:   
      correct.append(score / len(actual[i]))
    else:
      correct.append(1)
  return sum(correct) / len(actual) * 100
```

```python
print(accuracy_metric(df.Action_Verbs,action_verb))
print(accuracy_metric(df.Party_Obligate,party))
print(accuracy_metric(df.Party_Beneficiary,party))
print(accuracy_metric(df.Condition,condition))
```

```python
def f1(actual, predicted):
  precision = recall = f1 = []
  for i in range(len(actual)):
    tp = tn = fp = 0
    for j in predicted[i]:
      if j in actual[i]:
        tp += 1
      if j not in actual[i]:
        fp += 1
    for k in actual[i]:
      if k not in predicted[i]:
        tn += 1
    a = tp + fp
    b = tp + tn
    if a > 0:
      prec = tp / (tp +  fp)
      precision.append(prec)
    if b > 0:
      rec = tp / (tp + tn)
    c = prec + rec
    if c > 0:
      f1.append(2 * prec * rec / (prec + rec))
    #print(prec)

  return [sum(precision) / len(actual) * 100,sum(recall) / len(actual) * 100, sum(f1) / len(actual) * 100]
```

```python
print(f1(df.Action_Verbs,action_verb))
print(f1(df.Party_Obligate,party))
print(f1(df.Party_Beneficiary,party))
print(f1(df.Condition,condition))
```
